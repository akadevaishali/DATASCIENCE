```python
marks=[34,43,23,35,67,7,89,76,8]
```

```python
plt.boxplot(marks)
```

```python
q3=np.percentile(marks,75)
q1=np.percentile(marks,25)
```

```python
q3,q1
```

```python
iqr=q3-q1
iqr #interquartile range
```

```python
#lower bound
lb=q1-1.5 *  iqr
#upper bound
ub = q3 + 1.5 * iqr
```

```python
lb,ub
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
sns.jointplot(x = 'tip',y='total_bill',data=tips,hue='time')
plt.show()
```

```python
#HISTOGRAM
```

```python

plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'],kde=True)
plt.show()
```

```python
plt.title('Outliers in Tip')

plt.xlabel('Day')
plt.ylabel('Tip')
sns.boxenplot(x = 'day',y = 'tip',data = tips)
plt.show()

```

```python
sns.pairplot(tips,hue='smoker');
plt.show()
```
